```python
from sklearn.linear_model import Perceptron
from sklearn import datasets
from sklearn.model_selection import train_test_split# train laf ko tesst
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

iris=datasets.load_iris()
x=iris.data
y=iris.target
print(x[:5])
print(y[:5])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
sc=StandardScaler()
sc.fit(x_train)
x_train_std=sc.transform(x_train)
x_test_std=sc.transform(x_test)
print(x_train_std[:5],x_train[:5])
print(y[:5])

model=Perceptron(max_iter=40,eta0=0.1,random_state=0)  #(so lan hoc, toc do hoc, trang thai lay gia tri)
model.fit(x_train_std,y_train)
y_pred= model.predict(x_test_std)
print(y_pred)
print(y_test)
print('do chinh xac hinh anh',accuracy_score(y_test,y_pred))
```

```python
import pandas as pd

data = pd.read_csv('crabs.csv')
data['sex'].replace({"M": 1, "F": 0}, inplace=True)
data["speies"].replace({"B": 1, "O": 0}, inplace=True)
x=data
y=data['sex']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
sc=StandardScaler()
sc.fit(x_train)
x_train_std=sc.transform(x_train)
x_test_std=sc.transform(x_test)
print(x_train_std[:5],x_train[:5])
print(y[:5])

model=Perceptron(max_iter=40,eta0=0.1,random_state=0)  #(so lan hoc, toc do hoc, trang thai lay gia tri)
model.fit(x_train_std,y_train)
y_pred= model.predict(x_test_std)
print(y_pred)
print(y_test)
print('do chinh xac hinh anh',accuracy_score(y_test,y_pred))
```
