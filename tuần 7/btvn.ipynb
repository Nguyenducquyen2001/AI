```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np

url = 'lung_cancer_examples.csv'
data = pd.read_csv(url)
data.drop(columns=['Name','Surname'], inplace=True)
data.hist()# số lần lặp lại trục tung, trục hoành là giá trị 
plt.show
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot= True)
```

```python
X = pd.DataFrame({'x1': data['Alkhol'], 'x2': data['Alkhol']**5})

# Tạo mô hình hồi quy tuyến tính
model = linear_model.LinearRegression()
model.fit(X, y=data['Result'])

# In thông tin về mô hình
print('Intercept của mô hình:', model.intercept_)
print('Coefficient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y=data['Result']))

# Vẽ biểu đồ
plt.scatter(data['Alkhol'], data['Result'], color='red')
range_ = np.arange(0, 8, 0.05)
predictor = pd.DataFrame({'x1': range_, 'x2': range_ **5})
y_pred = model.predict(predictor)
plt.plot(range_, y_pred, color='blue')
plt.show()
```
