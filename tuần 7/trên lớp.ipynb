```python

import pandas as pd
import matplotlib.pyplot as plt

url = 'mtcars.csv'
data = pd.read_csv(url)
data.hist()# số lần lặp lại trục tung, trục hoành là giá trị 
plt.show
correlation = data.corr()
print(correlation)

```

```python
import seaborn as sns
sns.heatmap(correlation,annot= True)
#mpg phụ thuộc chính vào wt
```

```python
data.plot(kind='scatter', x='wt',y='mpg',figsize=(9,9),color= 'black')#ham ve
```

```python
from sklearn import linear_model

data.plot(kind='scatter', x='wt',y='mpg',figsize=(9,9),color= 'black')#ham ve
model= linear_model.LinearRegression()
model.fit(X= pd.DataFrame(data['wt']),y=data['mpg'])
print(model.intercept_)
print(model.coef_)
model.score(X= pd.DataFrame(data['wt']),y=data['mpg'])
```

```python
# vẽ hàm hồi quy bậc 1 
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
y_prad= model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_prad,color='blue')# đường đại diện cho mối quan hệ giữa wt và mpg 
```

```python
# hàm hồi quy bậc 2 
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T
model.fit(X, y = data['mpg'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X, y = data['mpg']))
range = np.arange(1.5,5.5,0.1)
data.plot(kind='scatter', x='wt',y='mpg',figsize=(9,9),color= 'red', xlim = (0,7))#ham ve
predictor = pd.DataFrame([range,range**2]).T
y_pred = model. predict(X = predictor)
plt.plot(range, y_pred, color ='blue')
```

```python
# hàm hồi quy bậc 3 
import numpy as np

model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**3]).T
model.fit(X, y = data['mpg'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X, y = data['mpg']))

range = np.arange(1.5,5.5,0.1)
data.plot(kind='scatter', x='wt',y='mpg',figsize=(9,9),color= 'red', xlim = (0,7))#ham ve
predictor = pd.DataFrame([range,range**3]).T
y_pred = model. predict(X = predictor)
plt.plot(range, y_pred, color ='blue')
```

```python
# hồi quy tuyến tính 
model = linear_model.LinearRegression()
X = data.loc[:,['wt','hp']]
y = data['mpg']
model.fit(X, y)
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X, y))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'Salary_Data.csv'
data = pd.read_csv(url)
data.hist()# số lần lặp lại trục tung, trục hoành là giá trị 
plt.show
correlation = data.corr()
print(correlation)
```

```python
# hàm hồi quy bậc 1 
from sklearn import linear_model

data.plot(kind='scatter', x='Salary',y='YearsExperience',figsize=(9,9),color= 'black')#ham ve
model= linear_model.LinearRegression()
model.fit(X= pd.DataFrame(data['Salary']),y=data['YearsExperience'])
print(model.intercept_)
print(model.coef_)
model.score(X= pd.DataFrame(data['Salary']),y=data['YearsExperience'])
y_prad= model.predict(X=pd.DataFrame(data['Salary']))
plt.plot(data['Salary'],y_prad,color='blue')
```

```python
# hàm hồi quy bậc 2 
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'], data['Salary']**10]).T
model.fit(X, y = data['YearsExperience'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X, y = data['YearsExperience']))
range = np.arange(1.5,5.5,0.1)
data.plot(kind='scatter', x='Salary',y='YearsExperience',figsize=(9,9),color= 'red')#ham ve
predictor = pd.DataFrame([range,range**10]).T
y_pred = model. predict(X = predictor)
plt.plot(range, y_pred, color ='blue')
```

```python
X = pd.DataFrame({'x1': data['Salary'], 'x2': data['Salary']**3})

# Tạo mô hình hồi quy tuyến tính
model = linear_model.LinearRegression()
model.fit(X, y=data['YearsExperience'])

# In thông tin về mô hình
print('Intercept của mô hình:', model.intercept_)
print('Coefficient của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y=data['YearsExperience']))

# Vẽ biểu đồ
plt.scatter(data['Salary'], data['YearsExperience'], color='red')
range_ = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame({'x1': range_, 'x2': range_ **3})
y_pred = model.predict(predictor)
plt.plot(range_, y_pred, color='blue')
plt.show()
```
