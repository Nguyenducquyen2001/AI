```python
from keras.datasets import cifar100
import matplotlib.pyplot as plt
(x_train, y_train), (x_test, y_test) = cifar100.load_data()
# for i in range(30):
#   plt.subplot(5,6,i+1)
#   plt.tight_layout()
#   plt.imshow(x_train[i])
#plt.show()
x_train = x_train.reshape(50000,32*32*3)
x_train = x_train.astype('float32')/255
x_test = x_test.reshape(10000,32*32*3)
x_test = x_test.astype('float32')/255

from keras.utils import to_categorical
y_test = to_categorical(y_test, 100)
y_train = to_categorical(y_train, 100)
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(1000, activation = 'relu', input_shape=(32*32*3,), name='hidden1'))
# model.add(Dense(5000, activation = 'relu', name='hidden2'))
# model.add(Dense(2000, activation = 'relu', name= 'hidden3'))
# model.add(Dense(1000, activation = 'relu', name='hidden4'))
# model.add(Dense(100, activation = 'relu', name='hidden5'))
model.add(Dense(100, activation = 'softmax', name='outputlayout'))
model.compile(optimizer="adam", loss="mse", metrics=["mae"])

# Just use `fit` as usual
model.fit(x_train, y_train, epochs=55)

# Save the model to a file
model.save('my_model.h5')
```
